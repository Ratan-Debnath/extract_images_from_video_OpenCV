This project target to extracting images from video.It is the extrction of meaningful information from videos.

```python
# Import libraries
import cv2
import os
```

```python
# Read the video from specified path
video_path = cv2.VideoCapture("Cafe_Extract_Info_OpenCV.mp4")
```

```python
try:
    # creating a folder named Data_extract_img_from_video
    if not os.path.exists("Data_extract_img_from_video"):
        os.makedir("Data_extract_img_from_video")
        
#if not created then raise error
except OSError:
    print("Error: Creating directory of data")
```

```python
#frame
currentframe = 0

while(True):
    # reading from frame
    ret,frame = video_path.read()
    
    if ret:
        #creating images
        name = "./Data_extract_img_from_video/frame" + str(currentframe) + ".jpg"
        print("Creating..." + name)
        
        #writing the extracted images
        cv2.imwrite(name, frame)
        # increasing counter, show how many frames are created
        currentframe +=1
    else:
        break
        
#Release all space and windows once done
video_path.release()
cv2.destoryAllWindows()
```
